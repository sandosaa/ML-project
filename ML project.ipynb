 5-Handling Outliers
 </h1>
 -> Calculate the IQR for each numerical column and identify outliers using the 1.5*IQR rule.



</h1>


<h1 style="color:#A0A0A0; font-size:18px; text-align:center; font-weight:bold; text-align:left;">

 Outliers Before Cleaning

</h1>

```python
# cspell:disable
numeric_cols = data.columns
plt.figure(figsize=(20,5))
plt.boxplot(data[numeric_cols].values, tick_labels=numeric_cols, vert=True ,patch_artist=True, boxprops=dict(facecolor='#8C1818'))
plt.title("Outliers Before Cleaning")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
numeric_cols = data.select_dtypes(include=np.number).columns

# 1-Count Outliers Before Replacement
Q1 = data[numeric_cols].quantile(0.25)
Q3 = data[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outlier_counts_before = {}
outlier_counts_after = {}

for col in numeric_cols:
    outlier_count_before = data[(data[col] < lower_bound[col]) | (data[col] > upper_bound[col])].shape[0]
    outlier_counts_before[col] = outlier_count_before

    data[col] = np.where((data[col] < lower_bound[col]) | (data[col] > upper_bound[col]), data[col].median(), data[col])

    outlier_count_after = data[(data[col] < lower_bound[col]) | (data[col] > upper_bound[col])].shape[0]
    outlier_counts_after[col] = outlier_count_after

# 4- Combine in DataFrame 
outlier_summary = pd.DataFrame({
    "Outliers Before": outlier_counts_before,
    "Outliers After": outlier_counts_after
})

outlier_summary
```



<h1 style="color:#A0A0A0; font-size:18px; text-align:center; font-weight:bold; text-align:left;">

 <h1 style="color:#D3D3D3; font-size:20px; font-weight:bold; text-align:left;">
Feature Importance in Other Models:
</h1>
<h1 style="color:#A0A0A0; font-size:18px; text-align:center; font-weight:bold; text-align:left;">


Some models can give some form of feature importance, but they are not ideal:

| Model                  | Why it's not suitable for Feature Importance?  |
|------------------------|-----------|
| Logistic Regression    | importanceâ€ comes from the coefficients, but these values depend heavily on scaling â†’ not reliable unless all features are standardized.   |
| SVM                    | It does not naturally provide feature importance (except in linear SVMs, and even then itâ€™s not very meaningful).    |
| Decision Tree          | 	It has feature importance, but a single tree can overfit, making the importance values unstable and not generalizable.   |
| KNN                    |It has no feature importance at all, because the algorithm is based on distances, not learned weights.    |


â­ Main Advantage of Random Forest
It reduces overfitting and gives more stable, accurate predictions.

Because Random Forest builds many decision trees and combines their results, it becomes:

More accurate

More stable

Less sensitive to noise

Less likely to overfit compared to a single Decision Tree

This is the number one advantage of Random Forest
</h1>

```python
#cspell:disable
rf = RandomForestClassifier()
rf.fit(X, Y)

fi = pd.DataFrame({
    "Feature": X.columns,
    "Importance": rf.feature_importances_
}).sort_values(by="Importance", ascending=False)

print("\n--> Feature Importance (Random Forest) ")
print(fi)
```




<h1 style="color:#A0A0A0; font-size:18px; text-align:center; font-weight:bold; text-align:left;">
Feature Importance (Random Forest)
</h1>

```python
# cspell:disable
# Add cumulative importance
fi_sorted = fi.sort_values('Importance', ascending=False).reset_index(drop=True)
fi_sorted['Cumulative'] = fi_sorted['Importance'].cumsum()

# Plot
plt.figure(figsize=(10,6))
sns.barplot(
    x="Importance",
    y="Feature",
    data=fi,
    palette=list(reversed(sns.color_palette("Reds", n_colors=len(fi)))),
    hue="Feature",
    legend=False
)
# Annotate cumulative values on bars
for i, cum in enumerate(fi_sorted['Cumulative']):
    plt.text(fi_sorted['Importance'].iloc[i] + 0.002, i, f"{cum:.2f}", va='center', fontsize=9)

plt.title('Feature Importance (Random Forest) with Cumulative Sum', fontsize=14)
plt.xlabel('Importance', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.tight_layout()
plt.show()

# Print: How many features needed to reach 80% cumulative importance
n_features_80 = (fi_sorted['Cumulative'] <= 0.80).sum()
print(f"\nFeatures needed to explain 80% importance: {n_features_80}")
print("\nTop features with cumulative importance:")
print(fi_sorted[['Feature', 'Importance', 'Cumulative']].head(n_features_80 + 1))
```

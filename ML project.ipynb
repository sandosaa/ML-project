 5-Handling Outliers
 </h1>
 -> Calculate the IQR for each numerical column and identify outliers using the 1.5*IQR rule.



</h1>

```python
numeric_cols = data.select_dtypes(include=np.number).columns

# 1-Count Outliers Before Replacement
Q1 = data[numeric_cols].quantile(0.25)
Q3 = data[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

outlier_counts_before = {}
for col in numeric_cols:
    lower_bound = Q1[col] - 1.5*IQR[col]
    upper_bound = Q3[col] + 1.5*IQR[col]
    count = data[(data[col] < lower_bound) | (data[col] > upper_bound)].shape[0]
    outlier_counts_before[col] = count

# 2-Replace Outliers with Median 
for col in numeric_cols:
    lower_bound = Q1[col] - 1.5*IQR[col]
    upper_bound = Q3[col] + 1.5*IQR[col]
    median = data[col].median()
    data[col] = np.where((data[col] < lower_bound) | (data[col] > upper_bound), median, data[col])

# 3-Count Outliers After Replacement 
outlier_counts_after = {}
for col in numeric_cols:
    lower_bound = Q1[col] - 1.5*IQR[col]
    upper_bound = Q3[col] + 1.5*IQR[col]
    count = data[(data[col] < lower_bound) | (data[col] > upper_bound)].shape[0]
    outlier_counts_after[col] = count

# 4- Combine in DataFrame 
outlier_summary = pd.DataFrame({
    "Outliers Before": outlier_counts_before,
    "Outliers After": outlier_counts_after
})

outlier_summary
```

<h1 style="color:#D3D3D3; font-size:20px; font-weight:bold; text-align:left;">
<span style="color:#800000;">3-</span> Algorithms (Classification Models)

<h1 style="color:#A0A0A0; font-size:18px; text-align:center; font-weight:bold; text-align:left;">

 ðŸ”˜Import Required Libraries

</h1>

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```


<h1 style="color:#D3D3D3; font-size:20px; font-weight:bold; text-align:left;">
1-Logistic Regression
</h1>  


<h1 style="color:#A0A0A0; font-size:18px; text-align:center; font-weight:bold; text-align:left;">

Note:

 Logistic Regression is a simple linear model used as a first baseline.

 
 It works well when features have linear relationships with the target.

</h1>

```python
log_reg = LogisticRegression(max_iter=2000)
log_reg.fit(X_train_scaled, y_train)
y_pred_lr = log_reg.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred_lr)
report_dict = classification_report(y_test, y_pred_lr, output_dict=True)
df_report = pd.DataFrame(report_dict).transpose()
df_report.loc['accuracy'] = [accuracy, np.nan, np.nan, np.nan]

cm = confusion_matrix(y_test, y_pred_lr)
df_cm = pd.DataFrame(cm, index=[f"Actual_{cls}" for cls in log_reg.classes_],
                     columns=[f"Pred_{cls}" for cls in log_reg.classes_])

print("->  Logistic Regression Accuracy:" )
print("   ",accuracy)
print("\nâœ… Classification Report + Accuracy:")
print(df_report)
print("\nâœ… Confusion Matrix:")
print(df_cm)
```

<hr style="border: 1px solid #800000;">




<h1 style="color:#D3D3D3; font-size:20px; font-weight:bold; text-align:left;">
2-Support Vector Machine (SVM)

<h1 style="color:#A0A0A0; font-size:18px; text-align:center; font-weight:bold; text-align:left;">

SVM is powerful for classification especially for non-linear boundaries.

Works best when the data is scaled (which you already did).

```python
svm_clf = SVC(kernel='rbf')
svm_clf.fit(X_train_scaled, y_train)
y_pred_svm = svm_clf.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred_svm)
report_dict = classification_report(y_test, y_pred_svm, output_dict=True)
df_report = pd.DataFrame(report_dict).transpose()
df_report.loc['accuracy'] = [accuracy, np.nan, np.nan, np.nan]

cm = confusion_matrix(y_test, y_pred_svm)
df_cm = pd.DataFrame(cm, index=[f"Actual_{cls}" for cls in svm_clf.classes_],
                     columns=[f"Pred_{cls}" for cls in svm_clf.classes_])

print("-> SVM Accuracy:" )
print("   ",accuracy)
print("\nâœ… Classification Report + Accuracy:")
print(df_report)
print("\nâœ… Confusion Matrix:")
print(df_cm)
```
